```python
start_number = int(input("Enter Your 1st Number: "))
last_number = int(input("Enter your 2nd number: "))
for i in range(start_number,last_number+1):
    for j in range(1,11):
        print(f"{start_number} x {j} = {start_number*j}")
    print("\t")
    start_number += 1
```

```python
first_number = int(input("Enter 1st number: "))
last_number = int(input("Enter 2nd number: "))
for j in range(1,11):
    for i in range(first_number,last_number+1):
        print(f"{i} x {j} = {i*j}",end="\t")
    print("\n")
```
